```python
#Visualisasi data untuk Matematika dan Fisika UTBK 2019
import seaborn as sns
sns.set()
plt.figure(figsize=(17, 10))
plt.subplot(2,2,1)
plt.hist(score['score_mat'],
         bins=50,
         facecolor='y',
         alpha=0.80)
plt.grid()
plt.xlabel('Nilai Matematika')
plt.ylabel('Jumlah Peserta')
plt.title('Histogram Sebaran Nilai Matematika')

plt.subplot(2,2,2)
plt.hist(score['score_fis'],
         bins=50,
         facecolor='b',
         alpha=0.75)
plt.grid()
plt.xlabel('Nilai Fisika')
plt.ylabel('Jumlah Peserta')
plt.title('Histogram Sebaran Nilai Fisika')
plt.show()
```

```python
#Visualisasi data Kimia dan Biologi UTBK 2019
sns.set()
plt.figure(figsize=(17, 10))
plt.subplot(2,2,1)
plt.hist(score['score_kim'],
         bins=50,
         facecolor='r',
         alpha=0.75)
plt.grid()
plt.xlabel('Nilai Kimia')
plt.ylabel('Jumlah Peserta')
plt.title('Histogram Sebaran Nilai kimia')

plt.subplot(2,2,2)
plt.hist(score['score_bio'],
         bins=50,
         facecolor='g',
         alpha=0.75)
plt.grid()
plt.xlabel('Nilai Biologi')
plt.ylabel('Jumlah Peserta')
plt.title('Histogram Sebaran Nilai Biologi')
plt.show()
```

```python
#read data
df =  pd.read_csv('majors.csv')
df
```

```python
#mengubah data id universitas dari int menjadi string 
df['id_university'] = df['id_university'].apply(str)
df.info()
```

```python
#data
value = df['type'].value_counts()

#label
label = ['Saintek', 'Soshum']

#plot
fig = plt.figure(figsize =(10, 7))
plt.pie(value, labels = label, autopct='%1.1f%%')
plt.title('Peminat SAINTEK/SOSHUM UTBK 2019')
 
#show plot
plt.show()
```

```python
df2 = df.groupby(['id_university','type'], as_index=False)['capacity'].agg('sum')
df3 = df2.sort_values(by=['capacity'], ascending=False )
df3
```

```python
#sortir data berdasarkan tipe ujian
d1 = df3.loc[df3['type'] == 'humanities' ][:10]
d2 = df3.loc[df3['type'] == 'science'][:10]
```

```python
#menentukan x dan y 
x = d1['id_university']
y = d1['capacity']

#title dan label
plt.title('10 Kampus dengan daya tampung SOSHUM terbanyak')
plt.xlabel('ID Universitas')
plt.ylabel('Daya Tampung')

plt.bar(x,y)

plt.show()
```
